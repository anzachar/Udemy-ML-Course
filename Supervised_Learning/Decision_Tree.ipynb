# Decision Trees #

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import graphviz
```

## Loading iris dataset ##

```python
iris=load_iris()
X=iris.data
y=iris.target
```

## Splitting the data to training and test set ##

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=1)
```

## Creating Decision Trees Classifier ##
### Here, we pick maximum depth of 3 ###

```python
clf=DecisionTreeClassifier(max_depth=3)
```

## Training the Classifier on the training data set ##

```python
clf.fit(X_train, y_train)
```

## Making predictions on the test data set ##

```python
y_pred=clf.predict(X_test)
```

## Accuracy Score ##

```python
accuracy=accuracy_score(y_test, y_pred)
```

```python
print("Accuracy of the model predictions: ", accuracy)
```
