```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
height=np.array([5.2, 5.7, 5.9, 6.0, 6.4])
weight=np.array([125, 140, 155, 160, 180])
```

```python
plt.scatter(height, weight)
plt.xlabel("Height (inches)")
plt.ylabel("Weight (pounds)")
plt.show()
```

## Creating best-fit line ##

```python
m, b=np.polyfit(height, weight, 1) #np.polyfit(ind_var, dep_var, deg_of_poly)
```

```python
plt.scatter(height, weight)
plt.plot(height, m*height +b, color='red')
plt.xlabel("Height (inches)")
plt.ylabel("Weight (pounds)")
plt.show()
```

## Making linear regression model ##

```python
new_height=5.8
pred_weight=m*new_height +b
print("predicted weight based on linear regression: ", pred_weight)
```

## Example with dataset ##

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('car_purchasing_data.csv')
data
```

### Splitting our data ###

Now we can split our data to features and labels. 
In this case we have 
- 5 features: gender, age,  debt, annual salary, networth
- 1 label: purchasing ability

```python
X=data[['gender', 'age', 'salary', 'debt', 'worth']] # features
y=data['amount'] # label
```

### Building the model (training) using LinearRegression ### 

```python
model=LinearRegression()
model.fit(X,y) # training the model using X and y
```

### Making Predictions ###

```python
new_data=np.array([[1,33,75500,15645,335845]])
# making prediction using our trained model for new_data
predicted_price=model.predict(new_data)
print("purchasing price based on the new data will be: ", predicted_price)
```
